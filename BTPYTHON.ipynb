```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Tải dữ liệu lên chương trình
file = ('Sale_Data_Mixed.xlsx')
df = pd.read_excel(file)
df.head(10)
```

```python
#Tạo bảng Pivot Table thống kê số lượng sản phẩm bán ra của các thương hiệu
pivot1 = pd.pivot_table (df, values='Quantity',index='ProductBrand',
                                    aggfunc=['sum'])
pivot1
```

```python
#Vẽ biểu đồ số lượng sản phẩm bán ra của các thương hiệu
values1 = df.groupby('ProductBrand').sum()['Quantity']
brand = ['Dell','HP','Lenovo']
colors = ['gold','blue','red']
plt.bar(x = brand, height=values1,color=colors,width=0.5)
plt.xlabel('ProductBrand',fontweight ='bold',fontsize = 13)
plt.ylabel('Quantity',fontweight ='bold',fontsize = 13)
plt.show()
```

```python
##Tạo bảng Pivot Table thống kê doanh thu của từng chi nhánh
df['Sales']=df['Quantity']*df['Price']
values2 = df.groupby('BranchName').sum()['Sales']
pivot2 = pd.pivot_table (df, values='Sales',index='BranchName',
                                    aggfunc=['sum'])
pivot2
```

```python
#Sắp xếp tên các chi nhánh theo thứ tự của values2
cities=[]
for city,sales in values2.items():
    cities.append(city)
```

```python
#Vẽ biểu đồ doanh thu của các chi nhánh
colors2 = ['gold','blue','red','pink','green']
plt.bar(x = cities, height=values2,color=colors2,width=0.55)
plt.xlabel('BranchName',fontweight ='bold',fontsize = 13)
plt.ylabel('Sales',fontweight ='bold',fontsize = 13)
plt.show()
```

```python
#Tạo bảng Pivot Table thống kê số lượng sản phẩm bán ra từ tháng 1 đến tháng 12
pivot3 = pd.pivot_table (df, values='Quantity',columns='MONTH',
                                    aggfunc=['sum'])
pivot3
```

```python
#Vẽ biểu đồ số lượng sản phẩm bán ra từ tháng 1 đến tháng 12
values3=df.groupby('MONTH').sum()['Quantity']
month = df['MONTH'].unique()
plt.plot(month,values3,color='lime')
plt.xticks(month)
plt.xlabel('Month',fontweight ='bold',fontsize = 13)
plt.ylabel('Quantity',fontweight ='bold',fontsize = 13)
plt.grid()
plt.show()
```

```python
#Tạo bảng Pivot Table thống kê số lượng sản phẩm của các thương hiệu được bán ra tại các chi nhánh
pivot4 = pd.pivot_table (df, values='Quantity',index='ProductBrand',columns='BranchName',
                                    aggfunc=['count'], margins=True)
pivot4
```

```python
#Vẽ biểu đồ số lượng sản phẩm của các thương hiệu được bán ra tại các chi nhánh
dell = [586,656,897,385,1102]
hp = [588,625,959,434,1054]
lenovo = [554,644,906,414,1098]
index = np.arange(5)
width=0.2
plt.bar(index,dell,width,color='red',label='Dell')
plt.bar(index + width,hp,width,color='blue',label='HP')
plt.bar(index + 2*width,lenovo,width,color='lime',label='Lenovo')
plt.xlabel('BranchName',fontweight ='bold',fontsize = 13)
plt.ylabel('Quantity',fontweight ='bold',fontsize = 13)
plt.xticks(index + width/2,cities)
plt.legend(loc='best')
plt.show()
```
